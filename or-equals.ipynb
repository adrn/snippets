```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Generate some fake data

```python
t = np.linspace(0,100,1000)
y = np.cos(t)
```

```python
mask = y > 0.8
mask |= y < -0.3

plt.plot(t[mask], y[mask], marker='o', linestyle='none')
plt.ylim(-1.1, 1.1)
```

This is equivalent to

```python
mask1 = y > 0.8
mask2 = y < -0.3
mask = mask1 | mask2

plt.plot(t[mask], y[mask], marker='o', linestyle='none')
plt.ylim(-1.1, 1.1)
```
