```python
from datetime import datetime

# Third-party
import astropy.time as time
import astropy.coordinates as coord
import astropy.units as u
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pytz
%matplotlib inline
```

```python
tz = pytz.timezone('US/Eastern')
times = time.Time(datetime.now(tz)) + np.linspace(-5., 5., 100)*u.hour
```

```python
target = coord.SkyCoord(ra=150.*u.degree, dec=41.*u.degree)
loc = coord.EarthLocation(lat=40*u.degree, lon=-100*u.degree)
target = target.transform_to(coord.AltAz(obstime=times, location=loc))
# ignore all these warnings..
```

```python
plt.plot(times.datetime, target.secz, marker=None)

ticks,ticklabels = plt.xticks()
plt.setp(plt.xticks()[1], rotation=45, ha='right');
```
