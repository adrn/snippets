```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

This is the "old way" or the matlab way to do subplots:

```python
x = np.random.random(size=1000)
y = np.random.random(size=1000)
z = np.random.random(size=1000)

plt.figure(figsize=(10,6))

plt.subplot(311)
plt.plot(x, y, linestyle='none')

plt.subplot(312)
plt.plot(x, z, linestyle='none')

plt.subplot(313)
plt.plot(y, z, linestyle='none')
```

But often, you want to be able to have axes share properties. The newer way to handle subplots is with this function, which returns a figure object, and an array of axes objects:

```python
fig,axes = plt.subplots(3, 1, figsize=(10,6))

axes[0].plot(x, y, linestyle='none')
axes[1].plot(x, z, linestyle='none')
axes[2].plot(y, z, linestyle='none')
```

Now you can specify that you want them all to share an x axis:

```python
fig,axes = plt.subplots(3, 1, figsize=(10,6), sharex=True)

axes[0].plot(x, y, linestyle='none')
axes[1].plot(x, z, linestyle='none')
axes[2].plot(y, z, linestyle='none')

# this will make sure things don't overlap
fig.tight_layout()
```
