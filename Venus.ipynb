```python
# Third-party
import astropy.coordinates as coord
import astropy.time as at
import astropy.units as u
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
venus = coord.SkyCoord(ra=6.00630*u.hourangle, dec=26.01205*u.deg)
```

```python
columbia = coord.EarthLocation(lat=40.8075*u.deg, lon=-73.9619*u.deg)
```

```python
utcoffset = -4 * u.hour # Eastern daylight time
times = at.Time('2015-05-08 20:00:00') - utcoffset
dt = np.linspace(0, 2, 100)*u.hour
```

```python
alt_az = venus.transform_to(coord.AltAz(obstime=times+dt, location=columbia))  
```

```python
plt.figure(figsize=(8,6))
plt.plot_date((times+dt+utcoffset).datetime, alt_az.alt.degree)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.ylim(14,41)

plt.title('Venus')
plt.xlabel('Time (NYC)')
plt.ylabel('Altitude [deg]')
```
