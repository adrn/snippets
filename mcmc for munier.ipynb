```python
import emcee
import matplotlib.pyplot as plt
import numpy as np
import triangle
%matplotlib inline
```

Generate some fake data...

```python
true_m = -3.
true_b = 21.

xdata = np.random.uniform(0,10,size=30)
ydata = true_m*xdata + true_b
sigma_y = 1.
ydata += np.random.normal(0,sigma_y,size=len(xdata))
```

Plot the fake data

```python
plt.plot(xdata, ydata, linestyle='none', marker='o')
```

Ok, no we forget about $\sigma_y$ because in reality you don't have that value. But that's fine, because we'll include it as a parameter to fit for. The assumptions this makes are:

1. The scatter is only in the y direction
2. The scatter is Gaussian

Dunno how bad of an assumption this is for your data...but makes the likelihood very simple:

```python
from scipy.stats import norm
```

```python
def log_likelihood(parameters, x, y):
    m,b,sigma_y = parameters
    
    model_y = m*x + b
    return -np.sum(np.log(sigma_y) + 0.5*((y - model_y)/sigma_y)**2)
    #return norm.logpdf(y, loc=model_y, scale=sigma_y).sum()
```

Now we can easily add priors on the parameters. I'll use a logarithmic prior on the variance, $\sigma_y$, for reasons explained here: http://en.wikipedia.org/wiki/Jeffreys_prior

```python
def log_prior(parameters):
    m,b,sigma_y = parameters
    
    # set bounds on the slope
    if m > 0 or m < -100.:
        return -np.inf
    
    # set bounds on the intercept
    if b > 1000 or b < 0.:
        return -np.inf
    
    if sigma_y <= 0.:
        return -np.inf
    
    # return -np.log(sigma_y)
    return 0.

def log_posterior(parameters, x, y):
    lp = log_prior(parameters)
    if not np.isfinite(lp):
        return -np.inf # short circuit if prior is -infinity
    
    return lp + log_likelihood(parameters, x, y)
```

Fire up the MCMC

```python
sampler = emcee.EnsembleSampler(nwalkers=16, dim=3, lnpostfn=log_posterior, 
                                args=(xdata,ydata))
```

Randomly draw some initial positions for the MCMC chains

```python
p0 = np.zeros((sampler.k,sampler.dim))
p0[:,0] = np.random.normal(-3, 0.1, size=sampler.k)
p0[:,1] = np.random.normal(21, 0.1, size=sampler.k)
p0[:,2] = np.random.normal(1., 0.01, size=sampler.k)
```

```python
pos,prob,state = sampler.run_mcmc(p0, 100) # burn i n
sampler.reset() # throw away burn in steps
pos,prob,state = sampler.run_mcmc(pos, 1000)
```

```python
fig = triangle.corner(sampler.flatchain,
                      extents=[(-6,0),(10,30),(0,2)],
                      truths=[true_m, true_b, sigma_y],
                      labels=["$m$", "$b$", r"$\sigma_y$"])
```

Now you can just draw samples from the markov chain, which are like samples from the posterior distribution

$$
p(m,b,\sigma_y \,|\, x,y)
$$

You can then get
$$
p(y_i \,|\, m,b,\sigma_y,x_i)
$$

But this might be way overkill, and makes a lot of assumptions...

```python
ix = np.random.randint(len(sampler.flatchain), size=10000)

x_i = 5.1 # or whatever, drawn from some other distribution
ys = np.zeros_like(ix)
for i,p in enumerate(sampler.flatchain[ix]):
    m,b,s = p
    ys[i] = m*x_i + b + np.random.normal(0.,s)
    
plt.hist(ys);
```
