```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,256,1024)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x, y1, c='#55aacc')
plt.plot(x, y2, c='#aa55cc')
```

```python
plt.plot(x, y1, c='#55aacc', lw=4.)
plt.plot(x, y2, c='#aa55cc', lw=4.)
```

```python
plt.plot(x, y1, c='#55aacc', marker=None, lw=4.)
plt.plot(x, y2, c='#aa55cc', marker=None, lw=4.)
```

```python
plt.plot(x, y1, color='#55aacc', marker=None, lw=4.)
plt.plot(x, y2, color='#aa55cc', marker=None, lw=4.)
```
